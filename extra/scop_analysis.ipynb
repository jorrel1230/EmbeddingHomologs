```python
import pandas as pd
```

```python
scop_df = pd.read_csv("/scratch/gpfs/jr8867/main/db/indices/baseline/baseline_metadata.csv")
```

```python
max_length = scop_df['seq_length'].max()
print(f"Max length sequence in 'seq': {max_length}")
```

```python
import matplotlib.pyplot as plt

# Extract lengths of sequences
scop_df['seq_length'] = scop_df['seq'].apply(len)

# Plot boxplot of sequence lengths
plt.figure(figsize=(10, 6))
plt.boxplot(scop_df['seq_length'], vert=False, patch_artist=True, boxprops=dict(facecolor='blue', alpha=0.7))
plt.title('Boxplot of Amino Acid Sequence Lengths')
plt.xlabel('Sequence Length')
plt.grid(axis='x', alpha=0.75)
plt.show()


# Extract lengths of sequences
scop_df['seq_length'] = scop_df['seq'].apply(lambda x : x[:min(len(x), 1024)]).apply(len)

# Plot boxplot of sequence lengths
plt.figure(figsize=(10, 6))
plt.boxplot(scop_df['seq_length'], vert=False, patch_artist=True, boxprops=dict(facecolor='blue', alpha=0.7))
plt.title('Boxplot of Amino Acid Sequence Lengths, After Truncating at 1024')
plt.xlabel('Sequence Length')
plt.grid(axis='x', alpha=0.75)
plt.show()
```
