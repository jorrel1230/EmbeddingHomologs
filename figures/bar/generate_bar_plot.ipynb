```python
import numpy as np
import matplotlib.pyplot as plt

# Dummy data for 5 samples
sample_names = ['BLAST', 'Vanilla', 'Triplets\nRandom', 'Triplets\nHard', 'N-Pair\nRandom']
auroc_scores = [0.8, 0.7, 0.84, 0.81, 0.9]  # Replace with your actual AUROC scores
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightseagreen']

# Create the bar plot
plt.figure(figsize=(6, 4))
plt.bar(sample_names, auroc_scores, color=colors, width=0.3, edgecolor='black', linewidth=1, alpha=0.8)

# Add labels and title
plt.xlabel('Implementation')
plt.ylabel('AUROC Score')
plt.title('AUROC Scores of Different Implementations')
plt.ylim(0, 1)  # Set y-axis limit to 0-1 for AUROC scores
plt.grid(axis='y', linestyle='--')

# Add score annotations on top of the bars
for i, score in enumerate(auroc_scores):
    plt.text(i, score + 0.01, f'{score:.2f}', ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.savefig('auroc_bar_plot.png')
plt.show()
```
